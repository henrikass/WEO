```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_table('WEOOct2020all.tsv')
```

```python
df.head(3).T
```

```python
df.info()
```

```python
df.shape
```

```python
per_capita_df = df[df['Subject Descriptor']=="Gross domestic product per capita, constant prices"]
```

```python
is_USD = df['Units'] == 'Purchasing power parity; 2017 international dollar'
```

```python
per_capita_df['Units'].unique()
```

```python
USD_df = per_capita_df[is_USD]
```

```python
USD_df.shape
```

```python
USD_df['Country'].is_unique
```

```python
USD_df['2019'] = USD_df['2019'].str.replace(',', '')
USD_df['2018'] = USD_df['2018'].str.replace(',', '')
```

```python
USD_df['2019'] = pd.to_numeric(USD_df['2019'], errors='coerce')
USD_df['2018'] = pd.to_numeric(USD_df['2018'], errors='coerce')
```

```python
is_null = USD_df['2019'].isnull()
```

```python
USD_df[is_null]
```

```python
USD_df = USD_df.dropna(subset=['2019'])
```

```python
USD_df.shape
```

```python
USD_df['2018'].isnull().sum()
```

```python
USD_df.dtypes
```

```python
USD_df['GDP grow 2019-2018'] = USD_df['2019'] - USD_df['2018']
```

```python
USD_df.head(3).T
```

```python
sorted_USD_GDP_per_capita = USD_df.sort_values(by='GDP grow 2019-2018', ascending=False)
```
